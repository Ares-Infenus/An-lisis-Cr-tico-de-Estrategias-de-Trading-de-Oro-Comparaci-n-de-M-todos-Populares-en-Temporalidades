```python

# Inicializar la aplicación Dash
app = dash.Dash(__name__)

# Función para crear gráficos de velas
def create_candlestick(df, title):
    fig = go.Figure(data=[go.Candlestick(x=df['DateTime'],
                    open=df['Open'],
                    high=df['High'],
                    low=df['Low'],
                    close=df['Close'])])

    fig.update_layout(
        title=title,
        xaxis_title='Fecha',
        yaxis_title='Precio',
        xaxis_rangeslider_visible=False,  # Ocultar el deslizador de rango
    )
    return fig

# Layout de la aplicación con pestañas
app.layout = html.Div([
    dcc.Tabs([
        dcc.Tab(label='Temporalidad 1H', children=[
            dcc.Graph(figure=create_candlestick(XAU_USD_1H, 'Gráfico de Velas - 1H'))
        ]),
        dcc.Tab(label='Temporalidad 4H', children=[
            dcc.Graph(figure=create_candlestick(XAU_USD_4H, 'Gráfico de Velas - 4H'))
        ]),
        dcc.Tab(label='Temporalidad 30M', children=[
            dcc.Graph(figure=create_candlestick(XAU_USD_30M, 'Gráfico de Velas - 30M'))
        ]),
        dcc.Tab(label='Temporalidad 15M', children=[
            dcc.Graph(figure=create_candlestick(XAU_USD_15M, 'Gráfico de Velas - 15M'))
        ]),
    ])
## Para ejecutar elimina esto #####])

# Ejecutar la aplicación
if __name__ == '__main__':
    app.run_server(debug=True)
```

```python
from googleapiclient.discovery import build
import pandas as pd

# Configuración de la API
api_key = 'AIzaSyAtbeJeDiRaODomhqsR4yqboq69vT9B15Y'  # Reemplaza con tu clave de API
youtube = build('youtube', 'v3', developerKey=api_key)

def obtener_videos_y_detalles(busqueda, max_results=50):
    videos = []
    search_response = youtube.search().list(
        q=busqueda,
        part='id,snippet',
        maxResults=max_results
    ).execute()

    for item in search_response['items']:
        if item['id']['kind'] == 'youtube#video':
            video_id = item['id']['videoId']
            video_title = item['snippet']['title']
            canal_title = item['snippet']['channelTitle']
            videos.append((video_id, video_title, canal_title))

    return videos

def obtener_detalles_video(video_id):
    video_response = youtube.videos().list(
        part='statistics,snippet',
        id=video_id
    ).execute()

    detalles = video_response['items'][0]
    vistas = detalles['statistics'].get('viewCount', 0)
    likes = detalles['statistics'].get('likeCount', 0)
    canal_id = detalles['snippet']['channelId']

    # Obtener información del canal
    canal_response = youtube.channels().list(
        part='statistics',
        id=canal_id
    ).execute()

    subscriptores = canal_response['items'][0]['statistics'].get('subscriberCount', 0)
    num_comentarios = detalles['statistics'].get('commentCount', 0)

    return {
        'vistas': vistas,
        'subscriptores': subscriptores,
        'num_comentarios': num_comentarios
    }

# Ejemplo de uso
busqueda = 'Las mejores estrategias de trading'
videos = obtener_videos_y_detalles(busqueda, max_results=100)

# Almacenar los resultados en un DataFrame
datos = []
for video_id, video_title, canal_title in videos:
    detalles = obtener_detalles_video(video_id)
    datos.append({
        'Titulo': video_title,
        'Nombre del Canal': canal_title,
        'Subcriptores': detalles['subscriptores'],
        'Vistas Totales': detalles['vistas'],
        'Numero de Comentarios': detalles['num_comentarios']
    })

# Crear un DataFrame de pandas
df = pd.DataFrame(datos)
# Opcional: Guardar en un archivo CSV
df.to_csv('Dataframe_Youtube.csv', index=False)
```
