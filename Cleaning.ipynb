```python
#----# Importamos librerias necesarias para la limpieza #----#
from math import pi
import pandas as pd
import matplotlib as mlp
import numpy as np
import plotly.graph_objects as go  # Esta libreria se usa para visualizar el grafico
import dash                        # Esta libreria se usa para ver varios graficos en un portal web
from dash import dcc, html         # Esta libreria se usa para ver varios graficos en un portal Web
#-----# Importamos los Datos #--------------#                                                                                               # Temporalidades
XAU_USD_4H = pd.read_csv(r'C:\Users\spinz\OneDrive\Desktop\Proyect article Scientist\Data\4 Horas\XAU_4h_data_2004_to_2024-09-20.csv')      # 4 Hora
XAU_USD_1H = pd.read_csv(r'C:\Users\spinz\OneDrive\Desktop\Proyect article Scientist\Data\1 Hora\XAU_1h_data_2004_to_2024-09-20.csv')       # 1 Hora
XAU_USD_15M = pd.read_csv(r'C:\Users\spinz\OneDrive\Desktop\Proyect article Scientist\Data\15 Minutos\XAU_15m_data_2004_to_2024-20-09.csv') # 15 Minutos
XAU_USD_30M = pd.read_csv(r'C:\Users\spinz\OneDrive\Desktop\Proyect article Scientist\Data\30 Minutos\XAU_30m_data_2004_to_2024-09-20.csv') # 30 Minutos
XAU_USD_5M = pd.read_csv(r'C:\Users\spinz\OneDrive\Desktop\Proyect article Scientist\Data\5 Minutos\XAU_5m_data_2004_to_2024-09-20.csv')    # 5 Minutos

#-------# Ciclo de acomodo de fecha automatico #-----------#
dfs = [XAU_USD_15M,XAU_USD_1H,XAU_USD_30M,XAU_USD_4H,XAU_USD_5M]

# Iterar sobre cada DataFrame en la lista
for df in dfs:                                                                              #| Nota 1:Transformando la columna Date como Datetime,
    # Unir columnas 'Date' y 'Time', convertir a datetime, y establecer como índice         #| La razon detras e esto es que es mas facil su lectur-
    df['DateTime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])                          #| a para graficar
    df.drop(columns=['Date', 'Time'], inplace=True)                                         #|
XAU_USD_15M
```

```python

# Inicializar la aplicación Dash
app = dash.Dash(__name__)

# Función para crear gráficos de velas
def create_candlestick(df, title):
    fig = go.Figure(data=[go.Candlestick(x=df['DateTime'],
                    open=df['Open'],
                    high=df['High'],
                    low=df['Low'],
                    close=df['Close'])])

    fig.update_layout(
        title=title,
        xaxis_title='Fecha',
        yaxis_title='Precio',
        xaxis_rangeslider_visible=False,  # Ocultar el deslizador de rango
    )
    return fig

# Layout de la aplicación con pestañas
app.layout = html.Div([
    dcc.Tabs([
        dcc.Tab(label='Temporalidad 1H', children=[
            dcc.Graph(figure=create_candlestick(XAU_USD_1H, 'Gráfico de Velas - 1H'))
        ]),
        dcc.Tab(label='Temporalidad 4H', children=[
            dcc.Graph(figure=create_candlestick(XAU_USD_4H, 'Gráfico de Velas - 4H'))
        ]),
        dcc.Tab(label='Temporalidad 30M', children=[
            dcc.Graph(figure=create_candlestick(XAU_USD_30M, 'Gráfico de Velas - 30M'))
        ]),
        dcc.Tab(label='Temporalidad 15M', children=[
            dcc.Graph(figure=create_candlestick(XAU_USD_15M, 'Gráfico de Velas - 15M'))
        ]),
    ])
])

# Ejecutar la aplicación
if __name__ == '__main__':
    app.run_server(debug=True)
```
